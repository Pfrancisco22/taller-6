```python
users = {
    "admin": {"username": "admin", "password": "admin123", "role": "admin"},
    "seller": {"username": "seller", "password": "seller123", "role": "vendedor"}
}

products = {
    "001": {"product_name": "Martillo", "buy_price": 5.0, "sell_price": 10.0, "stock": 50, "unit_type": "unidad"},
    "002": {"product_name": "Destornillador", "buy_price": 3.0, "sell_price": 7.0, "stock": 30, "unit_type": "unidad"}
}

sales = []
current_user = None

def login():
    global current_user
    username = input("Ingrese el nombre de usuario: ")
    password = input("Ingrese la contraseña: ")
    for user in users.values():
        if user["username"] == username and user["password"] == password:
            current_user = user
            return True
    print("Credenciales incorrectas.")
    return False

def logout():
    global current_user
    current_user = None
    print("Sesión cerrada.")

def create_user():
    username = input("Nuevo nombre de usuario: ")
    password = input("Nueva contraseña: ")
    role = input("Rol (admin o vendedor): ")
    users[username] = {"username": username, "password": password, "role": role}
    print(f"Usuario {username} creado con éxito.")

def add_update_product():
    product_code = input("Código del producto: ")
    product_name = input("Nombre del producto: ")
    buy_price = float(input("Precio de compra: "))
    sell_price = float(input("Precio de venta: "))
    stock = int(input("Cantidad en stock: "))
    unit_type = input("Unidad de medida: ")
    products[product_code] = {
        "product_name": product_name,
        "buy_price": buy_price,
        "sell_price": sell_price,
        "stock": stock,
        "unit_type": unit_type
    }
    print(f"Producto {product_name} añadido/actualizado.")

def list_products():
    for code, product in products.items():
        print(f"{code} - {product['product_name']}: Precio Venta: {product['sell_price']}, Stock: {product['stock']} {product['unit_type']}")

def search_product():
    code = input("Ingrese el código del producto: ")
    product = products.get(code)
    if product:
        print(f"{product['product_name']}: Precio Venta: {product['sell_price']}, Stock: {product['stock']} {product['unit_type']}")
    else:
        print("Producto no encontrado.")

def buy_product():
    code = input("Ingrese el código del producto: ")
    quantity = int(input("Cantidad a comprar: "))
    if code in products:
        products[code]["stock"] += quantity
        print(f"{quantity} unidades añadidas al inventario de {products[code]['product_name']}.")
    else:
        print("Producto no encontrado.")

def list_sales():
    for sale in sales:
        print(f"{sale['product']}: Cantidad vendida: {sale['quantity']}, Subtotal: {sale['subtotal']}")

def manage_admin():
    while current_user:
        print("\nOpciones de Administrador:")
        print("1. Registrar usuario")
        print("2. Agregar/Actualizar producto")
        print("3. Consultar inventario")
        print("4. Buscar producto")
        print("5. Comprar producto")
        print("6. Resumir ventas")
        print("7. Cerrar sesión")
        option = input("Seleccione una opción: ")

        if option == "1":
            create_user()
        elif option == "2":
            add_update_product()
        elif option == "3":
            list_products()
        elif option == "4":
            search_product()
        elif option == "5":
            buy_product()
        elif option == "6":
            list_sales()
        elif option == "7":
            logout()
            break
        else:
            print("Opción no válida.")

def sell_product():
    code = input("Ingrese el código del producto: ")
    quantity = int(input("Cantidad a vender: "))
    product = products.get(code)
    if product and product["stock"] >= quantity:
        subtotal = product["sell_price"] * quantity
        product["stock"] -= quantity
        sales.append({"product": product["product_name"], "quantity": quantity, "subtotal": subtotal})
        print(f"Venta registrada. Subtotal: {subtotal}")
    else:
        print("Stock insuficiente o producto no encontrado.")

def manage_seller():
    while current_user:
        print("\nOpciones de Vendedor:")
        print("1. Consultar inventario")
        print("2. Buscar producto")
        print("3. Vender producto")
        print("4. Cerrar sesión")
        option = input("Seleccione una opción: ")

        if option == "1":
            list_products()
        elif option == "2":
            search_product()
        elif option == "3":
            sell_product()
        elif option == "4":
            logout()
            break
        else:
            print("Opción no válida.")

def main():
    print("Bienvenido a Shaman Tools")
    while True:
        print("\n1. Login")
        print("2. Salir")
        choice = input("Seleccione una opción: ")

        if choice == "1":
            if login():
                if current_user["role"] == "admin":
                    manage_admin()
                elif current_user["role"] == "vendedor":
                    manage_seller()
            else:
                print("Error de autenticación.")
        elif choice == "2":
            print("Saliendo del programa.")
            break
        else:
            print("Opción no válida.")

main()


```
